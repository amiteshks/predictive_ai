```python
import pandas as pd
```

```python
from sklearn.datasets import load_diabetes
diabetes = load_diabetes()

df = pd.DataFrame(data=diabetes.data, columns=diabetes.feature_names)
df
```

```python
df
df.info()
df.describe()
df.corr()
```

```python
X = df.drop(columns='s6')
y = df['s6']

X,y
```

```python
from sklearn.model_selection import train_test_split

#Create Train and Test Data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size =0.2)
X_train
```

```python
#Ridge
from sklearn.linear_model import Ridge

ridge_model = Ridge(alpha = 0.1) # 0.01, 0.1, 0.25, 0.5, 1.0

ridge_model.fit(X_train, y_train)
ridge_model.coef_
```

```python
y_predicted = ridge_model.predict(X_test)
y_predicted
```

```python
from sklearn.metrics import mean_absolute_error, root_mean_squared_error, mean_squared_error, r2_score
ridge_mae = mean_absolute_error(y_predicted, y_test)
ridge_mse = mean_squared_error(y_predicted, y_test)
ridge_rmse = root_mean_squared_error(y_predicted, y_test)

ridge_mae, ridge_mse, ridge_rmse
```

```python
alpha_values = [0.01, 0.1, 0.25, 0.5, 1.0]
for alpha_value in alpha_values:

    ridge_model = Ridge(alpha = alpha_value)

    ridge_model.fit(X_train, y_train)
    # print (f"ridge_model.coef_: {ridge_model.coef_}")


    y_predicted = ridge_model.predict(X_test)

    ridge_mae = mean_absolute_error(y_predicted, y_test)
    ridge_mse = mean_squared_error(y_predicted, y_test)
    ridge_rmse = root_mean_squared_error(y_predicted, y_test)


    print (f"ridge_mae: {ridge_mae}, ridge_mse: {ridge_mse}, ridge_rmse: {ridge_rmse}")

```
